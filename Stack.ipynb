```python
class Stack:
    def __init__(self, arr = None):
        if arr is None:
            self.items = []
        else:
            self.items = arr
            
    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("Stack Underflow!")
    
    def see(self):
        return [i for i in self.items] 

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise Exception("Stack is empty nothing to peek") 

    def size(self):
        return len(self.items)
        
    def __iter__(self):
        return iter(self.items)

    def __str__(self):
        return str(self.items)

s2 = Stack([23,54,2,-3,4-4,-4])
s2.see()

s = Stack()
s.push(36)
s.push(40)
s.push(70)
s.push(10)
s.push(59)
s.peek()
s.pop()
s.pop()
s.pop()
s.pop()

s.size()
s.peek()
           
```

    <div>
        <strong>Problem:</strong> Use a stack to reverse a given string.
    </div>
    <div>
        <strong>Example:</strong>
        <pre>Input: "hello"

    <div class="problem-hint">

    </div>

```python
def reverse(s):
    stack = Stack()

    for i in s:
        stack.push(i)

    result = ''

    while not stack.is_empty():
        result += stack.pop()
    
    return result

print(reverse("hello"))
```

    <div>
        <strong>Problem:</strong> Check if a string containing only the characters (), {}, and [] is valid.
    </div>
    <div>
        A valid string has matching opening and closing brackets in the correct order.
    </div>
    <div class="problem-example">
        <strong>Example:</strong>
        <pre>Input: "({[]})"
Output: True

Input: "({[})"
Output: False</pre>
    </div>
    <div class="problem-hint">
        (Hint: Use a stack to match pairs of brackets.)
    </div>
</div>


```python
def valid_brackets(s):
    dict = {')':'(','}':'{',']':'['}

    stack = Stack()
    for chr in s:
        if chr in '({[':
            stack.push(chr)
        if chr in ']})':
            if stack.is_empty() or stack.pop() != dict[chr]:
                return False 

    return stack.is_empty()

print(valid_brackets("({[]})"))
print(valid_brackets("({[})"))
```

<div class="problem-container">
    <div class="problem-title">3. Next Greater Element</div>
    <div>
        <strong>Problem:</strong> For each element in an array, find the next greater element to the right. If no such element exists, return -1.
    </div>
    <div class="problem-example">
        <strong>Example:</strong>
        <pre>Input: [4, 5, 2, 10]
Output: [5, 10, 10, -1]</pre>
    </div>
    <div class="problem-hint">
        (Hint: Use a stack to keep track of indices.)
    </div>
</div>


```python
def next_greater_ele(arr):
    stack = Stack()

    nge = {}

    for num in arr:
        while not stack.is_empty() and num > stack.peek():
            nge[stack.pop()] = num 

        stack.push(num)

    while not stack.is_empty():
        nge[stack.pop()] = -1

    result = [nge.get(num, -1) for num in arr]
    return result 

arr = [4,6,2,6,4,29,49]

print(next_greater_ele(arr))
```

<div class="problem-container">
    <div class="problem-title">4. Implement a Min Stack</div>
    <div>
        <strong>Problem:</strong> Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.
    </div>
    <div>
        <strong>Operations:</strong>
        <pre>push(x) — Push element x onto the stack.
pop() — Remove the top element.
top() — Get the top element.
getMin() — Retrieve the minimum element in the stack.</pre>
    </div>
    <div class="problem-hint">
        (Hint: Use an auxiliary stack to store minimum values.)
    </div>
</div>


```python
class Stack_for_4:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self,item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            self.items.pop()
        else:
            raise Exception("Stack Is Empty!")

    def top(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return 0
        
    def see(self):
        return [i for i in self.items] 

    def min_val(self):
        if not self.is_empty():
            return min(self.items) 
        else:
            raise Exception("Stack Is Empty!")



s = Stack_for_4()
s.push(5)
s.push(234)
s.push(56)
s.push(-1)
s.push(-78)
s.see()
s.min_val()
```

    <div class="problem-title">5. Sort a Stack</div>
    <div>
        <strong>Problem:</strong> Sort a stack such that the smallest elements are on the top. You can use only one additional stack for temporary storage.
    </div>
    <div class="problem-example">
        <strong>Example:</strong>
        <pre>Input: [34, 3, 31, 98, 92, 23]
Output: [3, 23, 31, 34, 92, 98]</pre>
    </div>
    <div class="problem-hint">
        (Hint: Compare and move elements between stacks.)
    </div>
</div>


```python
def sort_stack(input_stack):
    temp_stack = Stack()

    while not input_stack.is_empty():
        curr = input_stack.pop()

        while not temp_stack.is_empty() and temp_stack.peek() > curr:
            input_stack.push(temp_stack.pop())

        temp_stack.push(curr)

    return temp_stack

input_stack = Stack()
for item in [34, 3, 31, 98, 92, 23]:
    input_stack.push(item)

print("Original Stack:", input_stack)
sorted_stack = sort_stack(input_stack)
print("Sorted Stack:", sorted_stack)
```

    <div>
        <strong>Problem:</strong> Write a program to reverse the elements of a stack using recursion. You are not allowed to use any additional data structures (like arrays or lists) to store the stack elements. The stack must be reversed in place.
    </div>
    <div class="problem-example">
        <strong>Example:</strong>
        <pre>Input: A stack with elements [1, 2, 3, 4] (top to bottom: 4, 3, 2, 1).
Output: A reversed stack [4, 3, 2, 1] (top to bottom: 1, 2, 3, 4).</pre>

<div class="problem-container">
    <div class="problem-title">7. Evaluate a Postfix Expression Using a Stack</div>
    <div>
        <strong>Problem:</strong> Write a program to evaluate a postfix expression (Reverse Polish Notation).
    </div>
    <div class="problem-example">
        <strong>Example:</strong>
        <pre>Input: 231*+9-
Output: -4</pre>
    </div>
</div>


```python
def evaluate_postfix(exp):
    s = Stack()

    for i in exp:
        if i.isdigit():
            s.push(i)
        else:
            b = int(s.pop())
            a = int(s.pop())

            if i == '+':
                s.push(a+b)
            elif i == '-':
                s.push(a-b)
            elif i == '*':
                s.push(a*b)
            elif i == '/':
                s.push(a/b)

    return s.pop()
            
expression = "231*+9-"
print("Postfix Expression:", expression)
print("Result:", evaluate_postfix(expression))          
            
```

<div class="problem-container">
    <div class="problem-title">8. Implement a Queue Using Stacks</div>
    <div>
        <strong>Problem:</strong> Implement a queue using two stacks.
    </div>
    <div>
        <strong>Operations:</strong> <code>enqueue(x)</code> and <code>dequeue()</code>.
    </div>
    <div class="problem-hint">
        (Hint: Use two stacks: one for enqueue operations and the other for dequeue operations.)
    </div>
</div>


```python
class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, item):
        self.stack1.append(item)

    def dequeue(self):
        if not self.stack2:  # Transfer elements from stack1 to stack2 if stack2 is empty
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        if not self.stack2:
            raise IndexError("Queue is empty!")
        return self.stack2.pop()


queue = QueueUsingStacks()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print(queue.dequeue())  # Output: 1
print(queue.dequeue())  # Output: 2
queue.enqueue(4)
print(queue.dequeue())  # Output: 3
print(queue.dequeue())  # Output: 4
```
